```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('/content/file.tsv',sep='\t',names=column_names)
```

```python
df.head()
```

```python
movie_titles=pd.read_csv('/content/Movie_Id_Titles.csv')
movie_titles.head()
```

```python
data=pd.merge(df,movie_titles,on='item_id')
data.head()
```

average rating of movie

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

ratings for each movie.recommend only having high rated movies

```python
data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(data.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(data.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
moviemat=data.pivot_table(index='user_id',columns='title',values='rating'
)
moviemat.head()
# moviemat['Star Wars (1977)']
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
starwar_ratings=moviemat['Star Wars (1977)']
liarliar_ratings=moviemat['Liar Liar (1997)']
starwar_ratings.head()
```

```python
similar_to_starwars=moviemat.corrwith(starwar_ratings)

corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
