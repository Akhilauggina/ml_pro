```python
columns_to_drop = ["Surname", "RowNumber", "CustomerId"]

x_train = x_train.drop(columns=columns_to_drop)
x_test = x_test.drop(columns=columns_to_drop)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Ensure x is correctly defined by explicitly dropping non-feature columns
x = df.drop(["Exited", "Surname", "RowNumber", "CustomerId"], axis=1)
y = df["Exited"]

# Re-run train-test split with the corrected x
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)

# Apply StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(11,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(optimizer="adam",metrics=["accuracy"],loss="binary_crossentropy")
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,batch_size=50,epochs=30,verbose=1,validation_split=0.2)
```
